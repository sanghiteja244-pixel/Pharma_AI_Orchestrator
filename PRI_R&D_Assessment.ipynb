```python


import pandas as pd

# 1. MOCK R&D DATABASE (In a real scenario, this would be your AI Agent's output)
# Note: Links provided are to official FDA/Regulatory landing pages for verification.
pharma_data = [
    {
        "Molecule": "Vildagliptin",
        "Target_Indication": "Type 2 Diabetes",
        "Risk_Factor": "DPP-4 Inhibition / Renal Safety",
        "Source_Verification": "https://www.accessdata.fda.gov/scripts/cder/daf/",
        "Status": "Approved (EU/India)"
    },
    {
        "Molecule": "SV-Target-X",
        "Target_Indication": "Neuroscience (CRAMS Project)",
        "Risk_Factor": "High Blood-Brain Barrier Permeability",
        "Source_Verification": "https://clinicaltrials.gov/",
        "Status": "Phase II"
    }
]

# 2. THE PRI LOGIC ENGINE
def run_pri_assessment(data_list):
    # Convert list to a Pandas DataFrame for structured analysis
    df = pd.DataFrame(data_list)

    # Logic: Flagging items for the R&D team to review based on status
    def assess_risk(status):
        if "Phase II" in status:
            return "HIGH PRIORITY (Monitor Closely)"
        return "Standard Monitoring"

    df['R&D_Priority'] = df['Status'].apply(assess_risk)
    return df

# 3. EXECUTION
assessment_report = run_pri_assessment(pharma_data)

# Displaying the report
print("--- PRI ENGINE: R&D RISK INTELLIGENCE REPORT ---")
print(assessment_report[['Molecule', 'Status', 'R&D_Priority', 'Source_Verification']])
```

```python

# 1. THE DATABASE (Expanded for R&D)
data = [
    {"Molecule": "Vildagliptin", "Status": "Approved (EU/India)", "Source": "https://www.accessdata.fda.gov/scripts/cder/daf/"},
    {"Molecule": "SV-Target-X", "Status": "Phase II", "Source": "https://clinicaltrials.gov/"},
    {"Molecule": "Amlodipine", "Status": "Generic", "Source": "https://www.accessdata.fda.gov/scripts/cder/ob/"},
    {"Molecule": "Remdesivir", "Status": "Emergency Use/Approved", "Source": "https://www.fda.gov/drugs/"}
]

df = pd.DataFrame(data)

# 2. ADDING THE PRI SEARCH FEATURE
def search_pri_engine():
    print("--- PHARMA RISK INTELLIGENCE (PRI) SEARCH ---")
    query = input("Enter Molecule Name to check Risk: ").strip()

    # Searching the dataframe (Case-insensitive)
    result = df[df['Molecule'].str.contains(query, case=False)]

    if not result.empty:
        print("\n[MATCH FOUND]")
        # Adding the Risk Priority logic on the fly
        priority = "HIGH PRIORITY" if "Phase" in result['Status'].values[0] else "Standard"

        print(f"Molecule: {result['Molecule'].values[0]}")
        print(f"Status: {result['Status'].values[0]}")
        print(f"Risk: {priority}")
        print(f"Source: {result['Source'].values[0]}")
    else:
        print("\n[!] No records found. Check spelling or database.")

# 3. RUN THE SEARCH
search_pri_engine()
```
